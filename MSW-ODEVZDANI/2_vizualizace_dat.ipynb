```python
import matplotlib.pyplot as plt
import pandas as pd
import csv

##########################################
# matematika skore

def matematika_skore():
    
    x = [] # skore
    y = [] # pocet
    
    with open("exams.csv", "r") as soubor:
        graf = csv.reader(soubor, delimiter = ",")             # znak oddelujici sloupce
        
        # ziskani hodnot z csv souboru
        zapsane_hodnoty = []
        for radek in graf:
            zapsane_hodnoty.append(radek[5])
            
        zpoctene_hodnoty = []
        dvojce = []
        
        for hodnota in zapsane_hodnoty:
            
            if hodnota not in zpoctene_hodnoty:
                
                zpoctene_hodnoty.append(hodnota)
                pocet = zapsane_hodnoty.count(hodnota)
                               
                               # skore   # pocet
                dvojce.append((hodnota, pocet))
        
            # skore  # pocet
        for hodnota, pocet in dvojce:
            x.append(hodnota)            # skore
            y.append(pocet)              # pocet
            
        plt.bar(x,y, color = "g", width = 0.70, label = "matematika")
        plt.xlabel("pocet bodu")
        plt.ylabel("pocet lidi se stejnym ohodnocenim")
        plt.title("vyobrazeni bodu")
        plt.legend()
        plt.show()
        
matematika_skore()
```

```python

def vzdelani():
    
    x = [] # forma vzdelani
    y = [] # pocet
    
    with open("exams.csv", "r") as soubor:
        graf = csv.reader(soubor, delimiter = ",")
        
        zapsane_hodnoty = []
        for radek in graf:
            zapsane_hodnoty.append(radek[2])
            
        zpoctene_hodnoty = []
        dvojce = []
        
        for hodnota in zapsane_hodnoty:
            
            if hodnota not in zpoctene_hodnoty:
                
                zpoctene_hodnoty.append(hodnota)
                pocet = zapsane_hodnoty.count(hodnota)
                               
                               # skore   # pocet
                dvojce.append((hodnota, pocet))
        
        for hodnota, pocet in dvojce:
            x.append(hodnota)            # forma vzdelani
            y.append(pocet)              # pocet
        
        
        plt.pie(y, labels = x)
        plt.title("vzdelani rodicu studentu")
        plt.show()
        
vzdelani()
```

```python

def rasa_vzdelani_rodicu_masters_degree():
    
    x = [] # rasa
    y = [] # prumer vzdelani rodicu
    
    with open("exams.csv", "r") as soubor:
        graf = csv.reader(soubor, delimiter = ",")
        
        # tvorba dvojce - rasa, vzdelani rodice
        
        zapsane_hodnoty = []
        for radek in graf:
            zapsane_hodnoty.append((radek[1],radek[2]))
                                       
        # zpracovani skupin
        for rasa, vzdelani_rodice in zapsane_hodnoty:
            if rasa not in x:
                x.append(rasa)
                y.append(zapsane_hodnoty.count((rasa, "master's degree")))
                
        print(x)
        print(y)
    
    plt.scatter(x, y, color = "g")
    plt.xticks()
    plt.xlabel("rasa")
    plt.ylabel("pocet rodicu s master's")
    plt.title("vyskyt masters degree u rodicu jednotlivych ras")
    plt.show()

rasa_vzdelani_rodicu_masters_degree()
```
