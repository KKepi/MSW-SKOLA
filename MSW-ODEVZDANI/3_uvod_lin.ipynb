```python
import numpy as np
import sympy
from sympy import Symbol
import scipy.linalg
from scipy.linalg import solve
import time
import matplotlib.pyplot as plt
sympy.init_printing()
```

```python
# 3 uvod do linearni algebry
print("\ngaussova iteracni metoda: ")
def gauss(A, B, x, n):
    L = np.tril(A)         # dolni troj forma
    U = A-L                # horni troj forma
    
    for i in range(n):
        x = np.dot(np.linalg.inv(L), B-np.dot(U,x))
    return x

def gauss_seidel(A,B,n,x):
    x = gauss(A,B,x,n)
    return solve(A,B)

gauss_seidel(np.array([[20, -2, 6],[-6, 12, 1], [4,-1, -14]]), np.array([-1,2,3]),5, [0,0,0])
```

```python
print("\n prima metoda:")
def prima_metoda(A,B):
    x = np.linalg.solve(A,B)
    return x

prima_metoda(np.array([[20, -2, 6],[-6, 12, 1], [4,-1, -14]]), np.array([-1,2,3]))
```

```python
###############################################################################################
# graf pro Gauss metodu
print("graf")

osa_x1 = []
osa_y2 = []

for i in range(100):
    if i > 2:
        # nahodne hodnoty matice
        A = np.random.rand(i,i)
        B = np.random.rand(i,1)
        velikost = i*i
        
        # cas pro vypocitani
        start = time.time()
        gauss_seidel(A,B,5,[ 0 for x in range(i)])
        konec = time.time()
        celkovy_cas = konec - start
        
        osa_x1.append(velikost)
        osa_y2.append(celkovy_cas)

plt.plot(osa_x1,osa_y2)
plt.xlabel("velikost matice")
plt.ylabel("uplynuty cas")
plt.title("pocitani pomoci Gauss metody")
plt.show()
```

```python
# graf pro primou metodu

osa_x2 = []
osa_y3 = []

for i in range(100):
    if i > 2:
        # nahodne hodnoty matice
        A = np.random.rand(i,i)
        B = np.random.rand(i,1)
        velikost1 = i*i
        
        # cas pro vypocitani
        start1 = time.time()
        prima_metoda(A,B)
        konec2 = time.time()
        celkovy_cas1 = konec2 - start1
        
        osa_x2.append(velikost1)
        osa_y3.append(celkovy_cas1)
        
plt.plot(osa_x2,osa_y3)
plt.xlabel("velikost matice")
plt.ylabel("uplynuty cas")
plt.title("pocitani pomoci prime metody")
plt.show()
```
